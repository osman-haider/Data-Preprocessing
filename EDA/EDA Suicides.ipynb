```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Suicides = pd.read_csv('Suicides in India 2001-2012.csv')
```

# First perform EDA on suicides data set

```python
Suicides.head()
```

```python
plt.figure(figsize=(20,15))
plt.plot(Suicides['Year'].values,Suicides['Total'].values)
```

```python
Suicides.info()
```

```python
suicides = Suicides.sample(10000)
```

```python
suicides.head()
```

```python
plt.figure(figsize=(20,15))
plt.pie(suicides['Total'],labels=suicides['Gender'],autopct='%0.1f%%')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.bar(suicides['Year'],suicides['Total'],color='black')
plt.xlabel("Year")
plt.ylabel("Total")
```

```python
sns.scatterplot(data=suicides,x='Year',y='Total')
```

```python
plt.figure(figsize=(12,8))
plt.bar(suicides['State'],suicides['Total'],color='black')
plt.xticks(rotation='vertical')
plt.xlabel("State")
plt.ylabel("Total")
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.bar(suicides['Gender'],suicides['Total'],color='black')
plt.xticks(rotation='vertical')
plt.xlabel("Gender")
plt.ylabel("Total")
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.bar(suicides['Age_group'],suicides['Total'],color='black')
plt.xticks(rotation='vertical')
plt.xlabel("Age group")
plt.ylabel("Total")
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.bar(suicides['Type_code'],suicides['Total'],color='black')
plt.xticks(rotation='vertical')
plt.xlabel("Caused")
plt.ylabel("Total")
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.bar(suicides['Type'],suicides['Total'],color='black')
plt.xticks(rotation='vertical')
plt.xlabel("Type")
plt.ylabel("Total")
plt.show()
```

```python
sns.relplot(data=suicides,x='Total',y='Year',kind='scatter')
```
