```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plants = pd.read_csv('Planets.csv')
```

```python
plants.head()
```

```python
plants['Name'].value_counts()
```

```python
plants['LIGHT-YEARS FROM EARTH'].value_counts()
```

```python
plt.figure(figsize=(20,15))
plt.plot(plants['DISCOVERY DATE'].values,plants['LIGHT-YEARS FROM EARTH'].values)
```

```python
plt.figure(figsize=(15,7))

plt.scatter(plants['DISCOVERY DATE'].values,plants['LIGHT-YEARS FROM EARTH'].values,c=plants['DISCOVERY DATE'])
```

```python
plt.figure(figsize=(12,8))
sns.relplot(data=plants,x='DISCOVERY DATE',y='LIGHT-YEARS FROM EARTH',kind='scatter')
```

```python
plt.figure(figsize=(15,15))
plt.bar(plants['Name'],plants['DISCOVERY DATE'],color='green')
plt.xticks(rotation='vertical')
plt.xlabel("Name of plante")
plt.ylabel("Year")
plt.show()
```
