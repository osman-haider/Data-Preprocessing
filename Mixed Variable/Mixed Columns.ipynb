```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['Family'] = df['SibSp'] + df['Parch']
```

```python
df.drop(columns=['SibSp','Parch'],inplace=True)
```

```python
df.head()
```

```python
df['Family'].replace(0,'A',inplace=True)
```

```python
df.head()
```

```python
new_df = df[['Survived','Cabin','Ticket','Family']]
```

```python
new_df.head()
```

```python
new_df.dropna(inplace=True)
```

```python
new_df.head()
```

```python
new_df['Family'].unique()
```

```python
# extract numrical values from family column
new_df['family_numric'] = pd.to_numeric(new_df['Family'],errors='coerce',downcast='integer')
```

```python
# extract catogorical values from family column
new_df['family_catigorical'] = np.where(new_df['family_numric'].isnull(),new_df['Family'],np.nan)
```

```python
new_df.head()
```

```python
new_df['Cabin'].unique()
```

```python
#  Cabin column
new_df['cabin_num'] = new_df['Cabin'].str.extract('(\d+)') # captuer numrical part
new_df['cabin_cat'] = new_df['Cabin'].str[0] # captuer catigorical part
```

```python
new_df.head()
```

```python
new_df['Ticket'].unique()
```

```python
# extract the last bit of ticket as number
new_df['ticket_num'] = new_df['Ticket'].apply(lambda s: s.split()[-1])
new_df['ticket_num'] = pd.to_numeric(new_df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
new_df['ticket_cat'] = new_df['Ticket'].apply(lambda s: s.split()[0])
new_df['ticket_cat'] = np.where(new_df['ticket_cat'].str.isdigit(), np.nan,
                                new_df['ticket_cat'])

new_df.head(20)
```
