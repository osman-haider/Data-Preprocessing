```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')[['Name','Age','Pclass','SibSp','Parch','Survived']]

df.head()
```

# 1. feature construction

```python
df.dropna(inplace=True)

df.head()
```

```python
df['Family_size'] = df['SibSp'] + df['Parch'] + 1

df.head()
```

#### now we can remove SinSp,Parch columns

```python
df.drop(columns=['SibSp','Parch'],inplace=True)
```
