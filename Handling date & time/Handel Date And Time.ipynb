```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('GlobalLandTemperaturesByMajorCity.csv',usecols=['dt','City'])
```

```python
df.head()
```

#### there is cloumn name called dt in this session we work on this column

```python
df.dtypes
```

```python
df['dt'] = pd.to_datetime(df['dt'])
df.info()
```

# extract year

```python
df['date_year'] = df['dt'].dt.year

df.sample(5)
```

# extract month

```python
df['date_month_no'] = df['dt'].dt.month

df.head()
```

# extract month name

```python
df['date_month_name'] = df['dt'].dt.month_name()
```

```python
df.head()
```

# extract days

```python
df['date_day'] = df['dt'].dt.day

df.head()
```

# Day of week

```python
df['date_dow'] = df['dt'].dt.dayofweek
df.head()
```

# day of week - name

```python
df['date_dow_name'] = df['dt'].dt.day_name()

df.head()
```

# is weekend?

```python
df['date_is_weekend'] = np.where(df['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)

df.head()
```
