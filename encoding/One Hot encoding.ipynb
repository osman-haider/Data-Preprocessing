# One Hot Encodeing Using Sklearn

```python
x=data[['age','gender','review','education']]
y=data['purchased']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_tarin,y_test=train_test_split(x,y,test_size=0.2,random_state=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first',sparse=False,dtype=np.int32)
```

```python
x_train_new=ohe.fit_transform(x_train[['gender']])
x_test_new=ohe.fit_transform(x_test[['gender']])
```
